```python
# Step 1: Drop duplicates.
# Step 2: Drop irrelevant columns.
# Step 3: Fix last_name column. Remove unnecessary characters. (... , /)
# Step 4: Reformat phone_number column. (000-00-0000)
# Step 5: Split address by comma(,). 
# Step 6: Standardize Yes and No values in paying_customer and do_not_contact columns. 
```

```python
import pandas as pd 
```

```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df = df.drop(columns=['Not_Useful_Column'])
```

```python
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')

df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x if pd.isna(x) else x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df['Phone_Number'] = df['Phone_Number'].fillna('')
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand=True)
df
```

```python
df['Paying Customer'] = (df['Paying Customer']
    .str.strip()
    .str.upper()
    .replace({'YES' : 'Yes', 'NO' : 'No', 'Y' : 'Yes', 'N' : 'No'})
) 

df['Paying Customer'] = df['Paying Customer'].replace('N/A', '')

df
```

```python
df['Do_Not_Contact'] = (df['Do_Not_Contact']
    .str.strip()
    .str.upper()
    .replace({'YES' : 'Yes', 'NO' : 'No', 'Y' : 'Yes', 'N' : 'No'})
) 

df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')
df
```

```python
# Filter out Yes values in Do_Not_Contact
# df = df.loc[df['Do_Not_Contact'] == 'No'].reset_index(drop=True)

df = df.loc[df['Do_Not_Contact'] == 'No']
df
```

```python
# Filter out empty strings in Phone_Number

df = df.loc[df['Phone_Number'].notna() & (df['Phone_Number'].str.strip() != '')]
df
```

```python
# Reset index 

df = df.reset_index(drop=True)
df
```

```python
# Export into excel file
df_cleaned = df.copy()  
df_cleaned.to_excel('df_cleaned.xlsx', index=False)
```
